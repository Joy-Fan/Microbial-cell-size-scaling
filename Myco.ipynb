```r
# Clear the data
rm(list=ls())

#Main dataframe
intr <- read.csv("C:/Users/FAN/Desktop/Data/Westoby/major-dimensions-1.0.0/data/madin_et_al/intracellular_organisms.csv", as.is=TRUE)
myco <- read.csv("C:/Users/FAN/Desktop/Data/Westoby/major-dimensions-1.0.0/data/madin_et_al/mycoplasma.csv", as.is=TRUE)
Data <- read.csv("C:/Users/FAN/Desktop/Data/TData_MB.csv", as.is=TRUE)
Data$intracellular <- NA

for(i in 1:nrow(intr)) {
  if(intr$phyl_level[i] == "order") {
    Data$intracellular[Data$ConOrder == intr$name[i]] <- "1"
  } else if (intr$phyl_level[i] == "genus") {
    Data$intracellular[Data$ConGenus == intr$name[i]] <- "1"
  } else if (intr$phyl_level[i] == "species") {
    Data$intracellular[Data$GenusSpecies == intr$name[i]] <- "1"
  }
}
rm(intr)

Data$mycoplasma <- NA
for(i in 1:nrow(myco)) {
  
  if(myco$phyl_level[i] == "order") {
    Data$mycoplasma[Data$ConOrder == myco$name[i]] <- "1"
  } else if (myco$phyl_level[i] == "genus") {
    Data$mycoplasma[Data$ConGenus == myco$name[i]] <- "1"
  } else if (myco$phyl_level[i] == "species") {
    Data$mycoplasma[Data$ConGenus == myco$name[i]] <- "1"
  }
}
rm(myco)
Intra<-subset(Data,Data$intracellular=="1")
Myco<-subset(Data,Data$mycoplasma=="1")
Intra
```
